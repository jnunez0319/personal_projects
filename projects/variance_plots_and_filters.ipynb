```python
import numpy as np
import matplotlib.pyplot as plt
```

### Draw 100 samples x = [x1, x2] from a 2-dimensional Gaussian distribution with mean [0, 0]. and identity covariance matrix. Plot them on a scatter plot (x1 vs. x2).

```python
mean = [0, 0]
cov_identity = [[1, 0], [0, 1]]

# Draw samples
samples = np.random.multivariate_normal(mean, cov_identity, 100)

# Plot the samples
plt.scatter(samples[:, 0], samples[:, 1], color='blue')
plt.title("Scatter Plot with Mean [0, 0] and Identity Covariance Matrix")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()
```

### How does the scatter plot change if the mean is [ 1, 1]?

```python
mean = [1, 1]
cov_identity = [[1, 0], [0, 1]]

# Draw samples
samples = np.random.multivariate_normal(mean, cov_identity, 100)

# Plot the samples
plt.scatter(samples[:, 0], samples[:, 1], color='blue')
plt.title("Scatter Plot with Mean [1, 1] and Identity Covariance Matrix")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()
```

### How does the scatter plot change if you double the variance of each component (x1 & x2)?

```python
mean = [0, 0]
cov_identity = [[2, 0], [0, 2]]

# Draw samples
samples = np.random.multivariate_normal(mean, cov_identity, 100)

# Plot the samples
plt.scatter(samples[:, 0], samples[:, 1], color='blue')
plt.title("Scatter Plot with Doubled Covarience")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()
```

### How does the scatter plot change if the covariance matrix changes to the following? X = [[1 0.5][0.5 1]]

```python
mean = [0, 0]
cov_identity = [[1, 0.5], [0.5, 1]]

# Draw samples
samples = np.random.multivariate_normal(mean, cov_identity, 100)

# Plot the samples
plt.scatter(samples[:, 0], samples[:, 1], color='blue')
plt.title("Scatter Plot with Set Covarience")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()
```

### How does the scatter plot change if the covariance matrix changes to the following? X = [[1 -0.5][-0.5 1]]

```python
mean = [0, 0]
cov_identity = [[1, -0.5], [-0.5, 1]]

# Draw samples
samples = np.random.multivariate_normal(mean, cov_identity, 100)

# Plot the samples
plt.scatter(samples[:, 0], samples[:, 1], color='blue')
plt.title("Scatter Plot with Negative Set Covarience")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()
```
