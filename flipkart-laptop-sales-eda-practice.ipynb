# Formatting Columns

```python
df = raw.drop(columns=['Product image', 'Link'])

df['Actual price'] = (df['Actual price']
                      .replace({r'[₹, ]': '', 'NIL': np.nan, '': np.nan}, regex=True)
                      .replace('', np.nan)
                      .astype(float)
                     )
df['Discount price'] = (df['Discount price']
                        .replace({r'[₹, ]': '', 'NIL': np.nan, '': np.nan}, regex=True)
                        .replace('', np.nan)
                        .astype(float)
                        )
df['Stars'] =  (df['Stars']
                .replace('', np.nan)
                .replace('NIL', np.nan)
                .astype(float)
               )
df['Rating'] = (df['Rating']
                .str.extract(r'(\d[\d,]*)')[0]
                .replace(',', '', regex=True)
                .replace('NIL', np.nan)
                .astype(float)
               )
df['Reviews'] = (df['Reviews']
                .str.extract(r'(\d[\d,]*)')[0]
                .replace(',', '', regex=True)
                .replace('NIL', np.nan)
                .astype(float)
                )
df
```

```python
import re

# Function to extract processor type
def extract_processor(description):
    match = re.search(r'(Intel Core \w+\d+ Processor|AMD Ryzen \d+ \w+ Core Processor)', description)
    return match.group(0) if match else None

# Function to extract RAM size (numeric) and type
def extract_ram(description):
    match = re.search(r'(\d+) GB (\w+ RAM)', description)
    return int(match.group(1)) if match else None, match.group(2) if match else None

# Function to extract operating system
def extract_os(description):
    match = re.search(r'(Windows \d+|Mac OS|Linux)', description)
    return match.group(0) if match else None

# Function to extract storage size (numeric) and type
def extract_storage(description):
    # Look for the second occurrence of storage information after RAM
    storage_part = description.split('RAM')[-1]
    match = re.search(r'(\d+) (GB|TB) (\w+)', storage_part)
    size = int(match.group(1)) if match else None
    if match and 'TB' in match.group(2):
        size *= 1024  # Convert TB to GB
    return size, "SSD" if match else None

# Function to extract display size in both cm and inch, and convert cm to inch if inch is not found
def extract_display_size(description):
    # Extract display size in cm directly after 'SSD'
    display_cm_match = re.search(r'SSD(\d+\.\d+)\s*cm', description)
    display_cm = float(display_cm_match.group(1)) if display_cm_match else None
    
    # Extract display size in inches within brackets or convert from cm
    display_inch_match = re.search(r'\((\d+\.\d+)\s*Inch\)', description)
    if display_inch_match:
        display_inch = float(display_inch_match.group(1))
    elif display_cm:
        display_inch = round(display_cm * 0.393701, 2)
    else:
        display_inch = None
    
    return display_cm, display_inch

# Function to extract warranty as a numeric value
def extract_warranty(description):
    match = re.search(r'(\d+) Year', description)
    return int(match.group(1)) if match else None

# Apply the functions to create new columns
df['Processor'] = df['Description'].apply(extract_processor)
df['RAM Size (GB)'], df['RAM Type'] = zip(*df['Description'].apply(extract_ram))
df['Operating System'] = df['Description'].apply(extract_os)
df['Storage Size (GB)'], df['Storage Type'] = zip(*df['Description'].apply(extract_storage))
df['Display Size (cm)'], df['Display Size (Inch)'] = zip(*df['Description'].apply(extract_display_size))
df['Warranty (Years)'] = df['Description'].apply(extract_warranty)
df['Brand'] = df['Product Name'].apply(lambda x: x.split()[0])

df
```
